```python
from scipy.spatial.distance import euclidean,cosine
```

```python
def distance(mid_1,mid_2):
    m1=data.loc[mid_1]
    m2=data.loc[mid_2]
    rp_dis=euclidean(m1[1:3],m2[1:3])
    gen_dis=cosine(m1[2:],m2[2:])
    return rp_dis+gen_dis
```

```python
distance(3,50)
```

```python
def similar_movie(mid,k=5):
    brother=[]
    for new_mid in data.index:
        if mid!=new_mid:
            d=distance(mid,new_mid)
            name=data.loc[new_mid]['movie_name']
            brother.append((d,name))
    brother.sort()
    return [name for d,name in brother[:k]]

```

```python
similar_movie(201,k=10)
```
